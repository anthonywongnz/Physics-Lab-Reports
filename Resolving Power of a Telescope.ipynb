# Experiment 232: Resolving Power of a Telescope
### Aim
To study the resolving power of a small telescope.


```python
from pylab import *
from matplotlib import *

a=array([0.000478, 0.000910, 0.001246, 0.001556])
delta=array([1.25e-3, 8.33e-4, 6.25e-4, 5.00e-4])
inversea=1/a

#print(delta)
#print(inversea)

plot(inversea, delta, linestyle="-", marker="x", color="green")

(m,b) = polyfit(inversea, delta, 1) 
line=m*inversea+b
plot(inversea,line,linestyle="--")

err=[1.43e-4, 7.25e-5, 4.69e-5, 3.44e-5]
#print(err)
errorbar(inversea,line,yerr=err, linestyle="None")

print("The wavelength of the lamp is: ", m, "m", sep="")

xlabel("$1/a$")
ylabel("$\delta$")

title("$\delta$ vs. $1/a$")

show()
```

```python
from pylab import *
from matplotlib import *

a=array([1.5, 2.0, 2.5, 3.0, 4.0, 4.0])
sep=array([1.13, 0.98, 0.82, 0.67, 0.63, 0.46])
inversea=1/a

#print(delta)
#print(inversea)

plot(inversea, sep, linestyle="-", marker="x", color="green")
 
(m,b) = polyfit(inversea, sep, 1 ) 
line=m*inversea+b
plot(inversea,line,linestyle="--")
print("The equation of the line of best fit is y = ", m, "x + ", b, sep="")

xlabel("$1/a$")
ylabel("$Seperation$")

title("Seperation vs. $1/a$ (For Circular Aperture)")

show()
```
