```python
import matplotlib.pyplot as plt
import numpy as np

#T = Room Temperature
v1=array([0.507,0.517,0.528,0.538,0.548,0.557,0.567,0.577,0.587,0.597,0.607,0.617,0.627])
I1=array([0.00000986,0.00001324,0.00001816,0.00002450,0.00003390,0.00004566,0.00006426,0.00009486,0.00013686,0.00020266,0.00029706,0.00042566,0.00063306])

#T = -100
v2=array([0.781,0.794,0.806,0.814,0.821,0.827,0.834,0.84,0.846,0.852,0.856,0.864,0.87])
I2=array([0.00000446,0.00000770,0.00001260,0.00001902,0.00002842,0.00004034,0.00005992,0.00008960,0.00013168,0.00019756,0.00029204,0.00043073,0.00062820])

#T = -70
v3=array([0.713,0.726,0.738,0.748,0.757,0.763,0.771,0.779,0.786,0.793,0.801,0.808,0.815])
I3=array([0.00000590,0.00000906,0.00001394,0.00002036,0.00002968,0.00004158,0.00006018,0.00009062,0.00013288,0.00019874,0.00029318,0.00043184,0.00062930])

#T = -40
v4=array([0.616,0.633,0.649,0.663,0.68,0.693,0.704,0.714,0.723,0.732,0.74,0.749,0.757])
I4=array([0.00000784,0.00001092,0.00001580,0.00002206,0.00003126,0.00004306,0.00006252,0.00009212,0.00013414,0.00019996,0.00029440,0.00043302,0.00063246])

#T = -10
v5=array([0.554,0.567,0.586,0.601,0.619,0.631,0.643,0.653,0.663,0.672,0.682,0.691,0.7])
I5=array([0.00000906,0.00001220,0.00001706,0.00002328,0.00003246,0.00004426,0.00006374,0.00009334,0.00013534,0.00020116,0.00029556,0.00043418,0.00063360])

plt.figure(figsize=(16,9))
#T= Room Temperature 22.2
(m,b) = np.polyfit(log(I1), v1, 1)
line = m*log(I1)+b
plt.plot(log(I1),line,linestyle="--",marker="",color="black",label="Room Temperature")
plt.plot(log(I1),v1,linestyle="",marker="x",color="black")
print("T=22.2, y=",m,"x+",b, sep="")

#T= -100
(m,b) = np.polyfit(log(I2), v2, 1)
plt.line = m*log(I2)+b
plt.plot(log(I2),line,linestyle="--",marker="",color="green",label="T=-100$\degree$C")
plt.plot(log(I2),v2,linestyle="",marker="x",color="green")
print("T=-100, y=",m,"x+",b, sep="")

#T= -70
(m,b) = np.polyfit(log(I3), v3, 1)
plt.line = m*log(I3)+b
plt.plot(log(I3),line,linestyle="--",marker="",color="blue",label="RT=-70$\degree$C")
plt.plot(log(I3),v3,linestyle="",marker="x",color="blue")
print("T=-70, y=",m,"x+",b, sep="")

#T= -40
(m,b) = np.polyfit(log(I4), v4, 1)
line = m*log(I4)+b
plt.plot(log(I4),line,linestyle="--",marker="",color="orange",label="T=-40$\degree$C")
plt.plot(log(I4),v4,linestyle="",marker="x",color="orange")
print("T=-40, y=",m,"x+",b, sep="")

#T= -10
(m,b) = np.polyfit(log(I5), v5, 1)
line = m*log(I5)+b
plt.plot(log(I5),line,linestyle="--",marker="",color="red",label="T=-10$\degree$C")
plt.plot(log(I5),v5,linestyle="",marker="x",color="red")
print("T=-10, y=",m,"x+",b, sep="")

plt.legend(loc = 4, prop={'size':11})

plt.xlabel("Voltage, $V$")
plt.ylabel("Log of Current, $ln(I), A$")

plt.title("Log Current Against Voltage")
plt.show()
```
